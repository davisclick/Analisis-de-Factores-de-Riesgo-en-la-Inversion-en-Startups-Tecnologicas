```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
```

```python
# Configura el path a tu ChromeDriver
CHROMEDRIVER_PATH = "C:\\CWebDrivers\\chromedriver.exe"  # <--- AJUSTA ESTA RUTA

# Configura Selenium (modo headless opcional)
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-blink-features=AutomationControlled")
service = Service(CHROMEDRIVER_PATH)
driver = webdriver.Chrome(service=service, options=options)

# Carga el archivo CSV
df = pd.read_csv("datasets/cuentas_startup_twitter.csv" , sep=";")

# Funcin para obtener followers de una cuenta de Twitter
def get_followers(twitter_url):
    if not isinstance(twitter_url, str):
        return None
    try:
        driver.get(twitter_url)
        time.sleep(25)  # Espera para cargar la pgina completamente

        verified_followers_link = driver.find_element(By.XPATH, "//a[contains(@href, '/verified_followers')]")
        number_span = verified_followers_link.find_element(By.XPATH, ".//span[1]/span")
        return number_span.text.strip()
    except Exception:
        return None

# Procesa cada fila
followers_list = []
for index, row in df.iterrows():
    twitter_url = row["twitter"]
    print(f"Procesando: {twitter_url}")
    followers = get_followers(twitter_url)
    followers_list.append(followers)

# Agrega la columna
df["followers"] = followers_list

# Guarda el nuevo archivo
df.to_csv("cuentas_startup_twitter_con_seguidores.csv", index=False)
print("Archivo guardado como 'x_accounts_with_followers.csv'.")

# Cierra el navegador
driver.quit()
```
